```r
library(survival)
library(ggplot2)
library(dplyr)
library(reshape2)
library(stringr)
```

```r
dir_fig = 'R_clogit/for_publish_fin/'
file_main = 'fragile_x_all/methylation_table_complete_hps.tsv'
df_main <- read.table(file_main, header = TRUE, sep = "\t")
df_main <- df_main[df_main$in_trios=='TRUE',]
print(dim(df_main))
head(df_main)

df_main_female = df_main[df_main$sex=='Female',]
print(dim(df_main_female))
head(df_main_female)

df_main_male = df_main[df_main$sex=='Male',]
print(dim(df_main_male))
head(df_main_male)
```

```r
### logistic regression: phenotype as function of CGG repeat length and FMR1 methylation fraction
print('female')
female_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl')
#female_model <- glm(female_formula, data = df_main_female, family = binomial)
female_model <- glm(Phenotype ~ mean_repeat_length + frac_hap_methyl, data = df_main_female, family = binomial)
print(summary(female_model))

print('++++++++++++++++++ write result table...')
df_out <- as.data.frame(summary(female_model)$coefficients)
rownames(df_out) <- c("intercept", "cgg_repeat_length", "frac_hap_methyl_fmr1")
# formating the columns with number of digits
for (col in colnames(df_out)) {
    df_out[col] = format(df_out[col], digits = 4, nsmall = 2)
}
print(df_out)
file_out = paste0(dir_fig, 'logit_model_asd_afo_cgg_methyl.tsv')
print(file_out)
write.table(df_out, file = file_out, sep = "\t", row.names = TRUE, quote = FALSE)

#print('male')
#male_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl')
#male_model <- glm(male_formula, data = df_main_male, family = binomial)
#print(summary(male_model))

#print('combined')
#formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl')
#model <- glm(formula, data = df_main, family = binomial)
#print(summary(model))
```

```r
### conditional logit: phenotype as function of CGG repeat length and FMR1 methylation fraction
print('female')
female_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + strata(fid)')
#female_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + PC1 + PC2 + PC3 + strata(fid)')
female_model <- clogit(female_formula, data = df_main_female)
print(summary(female_model))

#print('male')
#male_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + strata(fid)')
##male_formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + PC1 + PC2 + PC3 + strata(fid)')
#male_model <- clogit(male_formula, data = df_main_male)
#print(summary(male_model))

#print('combined')
#formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + strata(fid)')
##formula = as.formula('Phenotype ~ mean_repeat_length + frac_hap_methyl + PC1 + PC2 + PC3 + strata(fid)')
#model <- clogit(formula, data = df_main)
#print(summary(model))
```

```r
file_main = 'fragile_x_tables/methylation_table.tsv'
df_main <- read.table(file_main, header = TRUE, sep = "\t")
print(dim(df_main))
head(df_main)

df_main_female = df_main[df_main$sex==2,]
print(dim(df_main_female))
head(df_main_female)

df_main_male = df_main[df_main$sex==1,]
print(dim(df_main_male))
head(df_main_male)
```

```r
### phenotype as function of FMR1 repeat length
print('combined')
formula = as.formula('Phenotype ~ hp_rep_fmr1 + PC1 + PC2 + PC3 + strata(fid)')
#formula = as.formula('Phenotype ~ hp_rep_fmr1 + strata(fid)')
model <- clogit(formula, data = df_main)
print(summary(model))
    
print('female')
female_formula = as.formula('Phenotype ~ hp_rep_fmr1 + PC1 + PC2 + PC3 + strata(fid)')
#female_formula = as.formula('Phenotype ~ hp_rep_fmr1 + strata(fid)')
female_model <- clogit(female_formula, data = df_main_female)
print(summary(female_model))

print('male')
male_formula = as.formula('Phenotype ~ hp_rep_fmr1 + PC1 + PC2 + PC3 + strata(fid)')
#male_formula = as.formula('Phenotype ~ hp_rep_fmr1 + strata(fid)')
male_model <- clogit(male_formula, data = df_main_male)
print(summary(male_model))
```
