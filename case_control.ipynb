```python
### investigate meta data
meta_file = '/expanse/projects/sebat1/miladm/UCSD/LONG_READ_COHORT/REACH_sample_info_orig.tsv'
df_meta = pd.read_table(meta_file, sep='\t', header=0)

rename_samples_dict = {'REACH000236': 'REACH000236_PB', 
                       'REACH000436': 'REACH000436_PB', 
                       'REACH000530': 'REACH000530_PB', 
                       'REACH000531': 'REACH000531_ONT', 
                       'REACH000532': 'REACH000532_ONT'}
df_meta['Sample_ID'] = df_meta.apply(lambda row: rename_samples_dict[row['Sample_ID']] 
                                     if row['Sample_ID'] in rename_samples_dict else row['Sample_ID'], axis=1)
display(df_meta)
```

```python
file_psam = '/expanse/projects/sebat1/miladm/UCSD/LONG_READ_COHORT/process_IL_LR/HWE/REACH_LR_platform.psam'
df_p = pd.read_table(file_psam, sep='\t', header=None, names = ['famid', 'sample_id', 'dad', 'mom', 'sex', 'phen'])
display(df_p)

df_p = df_p.merge(df_meta[['Sample_ID', 'Rel_Proband', 'Affected', 'Diagnosis_LAB']], left_on='sample_id', right_on='Sample_ID', how='inner')
df_p = df_p.fillna('')
display(df_p)
df_p['Affected_asd'] = df_p.apply(lambda row: 'Yes' if (((row['Rel_Proband'] == 'Proband') and ('REACH000236' not in row['Sample_ID'])) or
                                                        ('Autism' in row['Diagnosis_LAB']) or 
                                                        ('autism' in row['Diagnosis_LAB']) or
                                                        ('ASD' in row['Diagnosis_LAB']) or
                                                        ('Autistic' in row['Diagnosis_LAB']) or
                                                        ("dev't delay" in row['Diagnosis_LAB']) or
                                                        ("dev'tl delay" in row['Diagnosis_LAB']) or
                                                        ('developmental delay' in row['Diagnosis_LAB']) or
                                                        ('Developmental delay' in row['Diagnosis_LAB']) or
                                                        ('Asperger' in row['Diagnosis_LAB']) or 
                                                        ('asperger' in row['Diagnosis_LAB']) or 
                                                        ('PDD-NOS' in row['Diagnosis_LAB'])) else 'No', axis=1)
# Speech             Delay
# speech           delay
# Speech/processing  delay
# epilepsy
# Epilepsy
display(df_p)
print('Affected:')
print(df_p['Affected'].value_counts())
print('Affected_asd:')
print(df_p['Affected_asd'].value_counts())
file_name = 'samples_labDiag_affected.tsv'
df_p.to_csv(file_name, sep='\t', index=False)
```
